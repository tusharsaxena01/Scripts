```python
import os
```

```python
class DuplicateManager:
    def __init__(self, path1, path2):
        self.path1 = path1
        self.path2 = path2
        self.files1 = self.getFilesList(self.path1)
        self.files2 = self.getFilesList(self.path2)
        self.duplicates = self.findDuplicates()
        
        # self.listDuplicates = lambda: self.duplicates

    def getFilesList(self, path) -> list:
        filesList = []
        for root, directories, files in os.walk(path):
            for file in files:
                # self.files.append(os.self.path.join(root, file))
                filesList.append(file)
        return filesList
    
    def findDuplicates(self) -> list:
        return list(set(self.files1).intersection(set(self.files2)))
    
    def listDuplicates(self) -> None:
        for duplicate in self.duplicates:
            print(duplicate)
    
    def removeDuplicates(self) -> None:
        for duplicate in self.duplicates:
            os.remove(os.path.join(self.path1, duplicate))
            print(f"Removed {os.path.join(self.path1, duplicate)}")
        print(f"Removed {len(self.duplicates)} files")
```

```python
path1 = r'D:\phone\Pictures\Screenshots'
path2 = r'D:\phone\Pictures\Screenshots - Copy'
df = DuplicateManager(path1, path2)
# df.findDuplicates()
print("1. List Duplicates\n2. Delete Duplicates\n3. List and Delete Duplicates\n4. Exit")
choice = int(input("Enter Your Choice: "))
boolean = True
while boolean:
    
    match choice:
        case 1:
            # print(df.findDuplicates())
            df.listDuplicates()
            print("Duplicates: ", len(df.duplicates))
        case 2:
            print("Deleting Duplicates")
            await df.removeDuplicates()
            print("Done")
            boolean = False
        case 3:
            print("Listing and Deleting Duplicates")
            await df.listDuplicates()
            print("Duplicates: ", len(df.duplicates))
            await df.removeDuplicates()
            boolean = False
        case 4:
            print("Exiting")
            boolean = False
```

```python
import os

def list_files(directory):
    """
    Recursively list all files in a directory and its subdirectories.
    
    Args:
        directory (str): The path to the directory.
        
    Returns:
        list: A list containing paths to all files.
    """
    files_list = []
    for root, directories, files in os.walk(directory):
        for file in files:
            # files_list.append(os.path.join(root, file))
            files_list.append(file)
    return files_list

# Example usage:
directory_path = r'D:\phone\Pictures\Screenshots'
files = list_files(directory_path)
for file in files:
    print(file)
```
